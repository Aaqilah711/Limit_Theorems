### Law of Large Numbers

```python
from scipy.stats import expon
import time 
from IPython import display
import matplotlib.pyplot as plt
import numpy as np
```

```python
# loc=0
scale=3 
tmean, tvar, tskew, tkurt = expon.stats(scale=scale,moments='mvsk')

plt.figure()
for n in range(1,500):
    expon_data = expon.rvs(scale=scale,size=n) 
    mean=((expon_data)**1).mean()
    
    plt.axhline(y=tmean,color = 'r', linestyle = '-')
    plt.plot(n,mean,'.')
    
    plt.tick_params(axis='x')
    plt.tick_params(axis='y')
    
    plt.xlabel("Size of sample")
    plt.ylabel("Mean")
    
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(.01)
```

#### Here the horizontal red line shows the population mean of the exponenetial data. The dots show the mean of the samples of different sizes. We see that as the size of the sample increases, the mean of the samples approaches the mean of the population. This demonstrates the law of large numbers.

### Central Limit Theorem for mean

```python
#mean of the exponential population
loc=0 
scale=5 
mean, var, skew, kurt = expon.stats(loc=loc, scale=scale,moments='mvsk') 
print("mean= ",mean)
```

```python
mean_bar =[]

for k in range(100000): #loops 100000 times over samples of size 100  each
    sample = expon.rvs(loc=loc, scale=scale,size=100) 
    
    mean_bar.append((sample).mean()) 
    
plt.figure()
plt.hist(mean_bar,bins=100)
plt.xlabel(r'$\bar{y}$')
plt.show()
```
